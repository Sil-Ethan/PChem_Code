```python
import numpy as np
from matplotlib import pyplot as plt
import sympy as sy
import math
```

```python
def entropy(G,H):
    Boltzmann = 1.38e-23
    return (Boltzmann * np.log(np.math.factorial(G) / (np.math.factorial(G - H) * np.math.factorial(H))))/H
```

```python
vectorized_entropy = np.vectorize(entropy)
H = np.arange(1, 101)  # Use arange to get an integer array
G = 100

plt.plot(H, vectorized_entropy(G, H))
plt.title('Entropy per Particle vs. Number of Particles')
plt.xlabel('Number of Particles')
plt.ylabel('Entropy per Particle')
```

```python
vectorized_entropy = np.vectorize(entropy)
H = np.arange(1, 101)  # Use arange to get an integer array
G = 100

plt.plot(H, vectorized_entropy(G, H)*H)
plt.title('Entropy per Particle vs. Number of Particles')
plt.xlabel('Number of Particles')
plt.ylabel('Entropy per Particle')
```
