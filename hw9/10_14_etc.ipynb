This is some starter code to get you moving on problems 10-14, 10-15, and 10-17, which reproduce some of the plots from the book associated with liquid/vapor compositional phase diagrams for an ideal solution (where the interactions between the two mixed liquids are the same as if the liquids were pure). The plot here more or less replicates what appears in the book on pp. 396-397.

The key to all of this is to get the Raoult's law expressions for each gas set up correctly: these relate the vapor pressure of each component to their mole fractions in the liquid. (Note for the future: if you tweak the Raoult's law behaviors to account for nonideal interactions then you can make this plot work for non-ideal solutions which have things like azeotropes, etc.)

```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
```

```python
P1pure = 20.9 #vapor pressure of pure 1-propanol
P2pure = 45.2 #vapor pressure of pure 2-propanol
def P1(x1):
  return x1*P1pure
def P2(x2):
  return x2*P2pure
def Ptotal(x1):
  return P1(x1)+P2(1-x1) #note here I am using 1-x1 rather than x2 so there is just one argument in Ptotal
def y1(x1): #fractional composition of 1-propanol in gas phase
  return P1(x1)/(P1(x1)+P2(1-x1))
```

```python
xforplot = np.arange(0,1,0.02)
#Ptotalforplot = []
#yforplot = []
Ptotalforplot=Ptotal(xforplot)
yforplot=y1(xforplot)
```

```python
plt.plot(xforplot,Ptotalforplot,label=r'pressure vs x')
plt.plot(yforplot,Ptotalforplot,label='pressure vs y') #note: "y" is the new x-axis for this second curve
plt.legend()
plt.xlabel('mole fraction of 1-propanol')
plt.ylabel('total vapor pressure')
```

note that the plot that I have made here is for the total pressure vs the composition of component 1; you can also try to make the same plot vs the composition of component 2 (which is just 1-component 1, or a reversal of the x-axis as it appears above and is what appears in Fig. 10-4 in the text). Now: do problem 10-14, which is plotting x1 vs y1 (the composition of the liquid vs gas phase as the mixture is increased from 0% 1-propanol to 100% 1-propanol).

```python
x2 = np.arange(0,1,0.02)
def x2_Ptotal(x2):
  return P1(1-x2)+P2(x2) #note here I am using 1-x1 rather than x2 so there is just one argument in Ptotal
def y2(x2): #fractional composition of 1-propanol in gas phase
  return P2(x2)/(x2_Ptotal(x2))

Ptotalforplot=x2_Ptotal(x2)
y2_plot=y2(x2)

plt.plot(x2,Ptotalforplot,label=r'pressure vs x')
plt.plot(y2_plot,Ptotalforplot,label='pressure vs y') #note: "y" is the new x-axis for this second curve
plt.legend()
plt.xlabel('mole fraction of 1-propanol')
plt.ylabel('total vapor pressure')
```

Then do 10-17, which asks you to do some algebraic rearrangement.  (You can do this in Python or do it by hand.)
image.png

```python
P1pure = 20.9 #vapor pressure of pure 1-propanol
P2pure = 45.2 #vapor pressure of pure 2-propanol
def P1(x1):
  return x1*P1pure
def P2(x2):
  return x2*P2pure
def y2(x2):
  return P2(x2)/(P1pure+x2*(P2pure-P1pure))
def P_total(y):
  return P1pure*P2pure/(P2pure-y*(P2pure-P1pure))

y_forplot = np.arange(0,1,0.02)
P_totalforplot=P_total(y_forplot)

plt.plot(xforplot,Ptotalforplot,label=r'pressure vs x')
plt.plot(yforplot,Ptotalforplot,label='pressure vs y') #note: "y" is the new x-axis for this second curve
plt.legend()
plt.xlabel('y2')
plt.ylabel('total vapor pressure')
```

We can see below how our graph goes over the lower curve! Showing how with these adjustments, we get the same lower curve as before.
