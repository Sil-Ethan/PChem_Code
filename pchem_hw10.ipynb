```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants
```

```python
def reaction_change(K, P):
    return (K/(K+4*P))**0.5
```

```python
P_values = np.linspace(0.1,30,30)
K_v = 6.1
reaction_values = reaction_change(K_v, P_values)
```

```python
plt.scatter(P_values, reaction_values, label='extent of reaction')
plt.xlabel('Pressure')
plt.ylabel('Extent of Reaction')
plt.legend()
```

```python
def reaction_change_2(n, K, P):
    return -n/2 + 0.5 *((n**2-4*(P*n**2-K*n-K)/(K+(4*P))))**0.5 
```

```python
n1 = 0.5
n2 = 1
n3 = 2

reaction_plot1 = reaction_change_2(n1,K_v,P_values)
reaction_plot2 = reaction_change_2(n2,K_v,P_values)
reaction_plot3 = reaction_change_2(n3,K_v,P_values)
```

```python
plt.scatter(P_values, reaction_plot1, label='N1/N0 = 0.5')
plt.scatter(P_values, reaction_plot2, label='N1/N0 = 1.0')
plt.scatter(P_values, reaction_plot3, label='N1/N0 = 2.0')
plt.xlabel('Pressure')
plt.ylabel('Extent of Reaction')
plt.legend()
```
